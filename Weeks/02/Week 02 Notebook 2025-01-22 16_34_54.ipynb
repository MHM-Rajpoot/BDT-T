```
def count_words_in_file(file_path):
    """
    Counts the total number of words in a file.

    Args:
        file_path (str): Path to the file.

    Returns:
        int: Total number of words in the file.
    """
    try:
        with open(file_path, "r") as file:
            words = file.read().split()
            return len(words)
    except FileNotFoundError:
        print(f"File {file_path} not found.")
        return None

# Example usage
filepath = "/tmp/sample_log.txt"
total_words = count_words_in_file(filepath)
print(f"Total words in file: {total_words}")
```

```
def count_successful_requests(file_path):
    """
    Counts the number of successful HTTP requests (200 status code) in the log file.

    Args:
        file_path (str): Path to the log file.

    Returns:
        int: Number of successful requests.
    """
    try:
        with open(file_path, "r") as file:
            successful_requests = sum(1 for line in file if ' 200 ' in line)
            return successful_requests
    except FileNotFoundError:
        print(f"File {file_path} not found.")
        return None

# Example usage
filepath = "/tmp/sample_log.txt"
successful_requests = count_successful_requests(filepath)
print(f"Successful requests: {successful_requests}")
```
