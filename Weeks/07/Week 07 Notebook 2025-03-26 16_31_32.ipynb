```
import mlflow
import logging

logging.getLogger("mlflow") .setLevel (logging. ERROR)
mlflow.pyspark.ml. autolog()
```

```
occupancyDF = spark.read.csv("/FileStore/tables/Occupancy_Detection_Data.csv", header = "true", inferSchema="true")
```

### Challenge 1

```
occupancyDF.display()
```

```
occupancyDF.createOrReplaceTempView("occupancyView")
```

```
%sql
SELECT Occupancy, AVG(Temperature) AS Average_Temperature,
       AVG(Humidity) AS Average_Humidity,
       AVG(CO2) AS Average_CO2,
       AVG(HumidityRatio) AS Average_Humidity_Ratio
FROM occupancyView
GROUP BY Occupancy
```

```
%sql
SELECT Occupancy,
       COUNT(*) AS Row_Count,
       MIN(Temperature) AS Min_Temperature,
       MAX(Temperature) AS Max_Temperature,
       MIN(Humidity) AS Min_Humidity,
       MAX(Humidity) AS Max_Humidity,
       MIN(CO2) AS Min_CO2,
       MAX(CO2) AS Max_CO2,
       MIN(HumidityRatio) AS Min_Humidity_Ratio,
       MAX(HumidityRatio) AS Max_Humidity_Ratio
FROM occupancyView
GROUP BY Occupancy
```

```
occupancyDF = occupancyDF.drop(occupancyDF. Timestamp)
occupancyDF . printSchema ()
```

```
from pyspark.ml. feature import RFormula

preprocess = RFormula ( formula="Occupancy ~ . ")
occupancyDF = preprocess. fit(occupancyDF) . transform(occupancyDF)

occupancyDF . show (5)
```

```
(trainingDF, testDF) = occupancyDF.randomSplit([0.7, 0.3], seed=100)
```

**View Model Parameters**
To view all parameters while training the model, run this cell and check the logs.

```
from pyspark.ml. classification import DecisionTreeClassifier

dt = DecisionTreeClassifier(labelCol="label", featuresCol="features")
model = dt. fit(trainingDF)
```

```
predictions = model. transform(testDF)
predictions. show()
```

```
from pyspark.ml. evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator (labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator. evaluate (predictions)

print("Accuracy = %g " % (accuracy))
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.functions import col
from pyspark.mllib.evaluation import MulticlassMetrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

predictionAndLabels = predictions.select(col("prediction"), col("label")).rdd.map(tuple)

metrics = MulticlassMetrics(predictionAndLabels)

conf_matrix = metrics.confusionMatrix().toArray()

labels = ["Unoccupied (0)", "Occupied (1)"]

plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt=".0f", cmap="Blues", xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```
from pyspark.ml. tuning import ParamGridBuilder

parameters = ParamGridBuilder ()\
.addGrid(dt. impurity, ["gini", "entropy"]) \
.addGrid (dt. maxDepth, [3, 5, 7])\
.addGrid(dt.maxBins, [16, 32, 64])\
.build()
```

```
from pyspark.ml. tuning import TrainValidationSplit

tvs = TrainValidationSplit()\
. setSeed (100) \
. setTrainRatio(0.75)\
. setEstimatorParamMaps (parameters) \
. setEstimator (dt) \
. setEvaluator (evaluator)
```

```
gridsearchModel = tvs. fit(trainingDF)
```

```
bestModel = gridsearchModel.bestModel
print("Parameters for the best model:")
print("MaxDepth Parameter: %g" %bestModel. getMaxDepth())
print("Impurity Parameter: %s" %bestModel. getImpurity())
print("MaxBins Parameter: %g" %bestModel. getMaxBins())
```

```
evaluator. evaluate(bestModel. transform(testDF))
```
