```
%python

from pyspark.sql.types import * 

mySchema = StructType ([ 
    StructField("index", IntegerType()) , 
    StructField("webpage", StringType()) , 
    StructField("associated_files", StringType())]) 

myRDD = sc.textFile("/FileStore/tables/webpage/*").map(lambda line: line.split("\t")).map(lambda values: [int(values[0]) , values[1] , values[2]]) 

webpages1 = spark.createDataFrame(myRDD , mySchema)

webpages1.show(5) 
```

### Challenge 1

```
%sql
SELECT 
    COUNT(DISTINCT CASE WHEN split_af LIKE '%.jpg' THEN webpage END) AS jpg_count,
    COUNT(DISTINCT CASE WHEN split_af LIKE '%.js' THEN webpage END) AS js_count,
    COUNT(DISTINCT CASE WHEN split_af LIKE '%.css' THEN webpage END) AS css_count
FROM explode_webpages;
```

```
%sql
CREATE OR REPLACE TEMP VIEW web_js AS SELECT webpage, split_af FROM explode_webpages WHERE split_af LIKE '%.js';
```

```
%sql
SELECT
    CASE
        WHEN pct_fema_sfha > 0 THEN 'within FEMA SFHA'
        ELSE 'outside FEMA SFHA'
    END AS region,
    AVG(
        CASE
            WHEN pct_fema_sfha > 0 THEN avg_risk_score_sfha
            ELSE avg_risk_score_no_sfha
        END
    ) AS avg_flood_risk_score
FROM zip_level_risk
GROUP BY region;
```

```
%sql
SELECT
    'within FEMA SFHA' AS region,
    AVG(avg_risk_score_sfha) AS avg_flood_risk_score
FROM zip_level_risk
WHERE pct_fema_sfha > 0
UNION ALL
SELECT
    'outside FEMA SFHA' AS region,
    AVG(avg_risk_score_no_sfha) AS avg_flood_risk_score
FROM zip_level_risk
WHERE pct_fema_sfha = 0;
```

```
%sql
SELECT
    zipcode,
    count_property,
    ROUND(count_floodfactor10 * 100.0 / count_property, 1) AS pct_risk_factor_10
FROM zip_level_risk
ORDER BY pct_risk_factor_10 DESC
LIMIT 10;
```
