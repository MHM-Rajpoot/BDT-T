```python
import pandas as pd

try:
  df = pd.read_csv('financial2023.csv', delimiter=',')
  print("Delimiter is comma (,)")
except pd.errors.EmptyDataError:
  print("File is empty")
except pd.errors.ParserError:
  try:
    df = pd.read_csv('financial2023.csv', delimiter='\t')
    print("Delimiter is tab (\t)")
  except pd.errors.EmptyDataError:
    print("File is empty")
  except pd.errors.ParserError:
    print("Unable to parse file with comma or tab delimiter")
```

```python
import json

with open('QC1.json', 'r') as f:
  data = json.load(f)

first_level_keys = list(data.keys())
print("First level keys:", first_level_keys)
```

```python
def count_nested_dicts(data, level=1):
  count = 0
  for k, v in data.items():
    if isinstance(v, dict):
      count += 1
      count += count_nested_dicts(v, level + 1)
  return count

nested_dict_count = count_nested_dicts(data)
print("Nested dictionaries:", nested_dict_count)
```
